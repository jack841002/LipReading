```python
import os 
import numpy as np
import cv2
from glob import glob
from PIL import Image
import imutils
import gc
```

# s1

```python
lists = glob("/home/eagleuser/Users/dragon/LipReading/GridCorpus/s1/Lip_image/*")
lists.sort()
```

```python
image_list_s1 = []
txt_list_s1 = []

for folder_path in lists:
    folder = folder_path.split('/')[-1]
    lists1 = glob("/home/eagleuser/Users/dragon/LipReading/GridCorpus/s1/Lip_image/" + folder + "/*")
    lists1.sort()
    if len(lists1) == 75:
        f = open('/home/eagleuser/Users/dragon/LipReading/GridCorpus/s1/text/' + folder + ".txt")
        for line in f:
            txt_list_s1.append(line)
    
    for img_path in lists1:        
        if len(lists1) == 75:
            image = Image.open(img_path)

            image = image.resize((100,50), Image.BILINEAR) 
            np_image = np.array(image)
            image_list_s1.append(np_image)
```

```python
del lists, folder, lists1, f, image, np_image
gc.collect()
```

```python
image_list_s1 = np.array(image_list_s1)
image_list_s1.shape    # (73800, 50, 100, 3)
print(len(txt_list_s1))       # 984
image_list_s1 = image_list_s1.reshape(( int(len(image_list_s1)/75), 75, 50, 100, 3))
print(image_list_s1.shape)    # (984, 75, 50, 100, 3)
```
