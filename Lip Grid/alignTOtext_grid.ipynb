```python
import os
import shutil
from os import listdir
from os.path import isfile, isdir, join
```

```python
# 指定要列出所有檔案的目錄
mypath = "./GridCorpus/s34/align/"                                    # modify

# 取得所有檔案名稱
files = listdir(mypath)
```

```python
files
```

```python
for index in files:
    path = mypath + index
    
    f = open(path, 'r')
    file = f.read()

    f_line = file.split("\n")
    del f_line[-1]  # delete last element ('')
    temp = ""
    for ff in f_line:
        temp = temp + ff.split(" ")[2] + " "
    temp = temp.split(' ')[1:-2]    

    # list to string
    sentence = " ".join(temp)
    
    name = "./GridCorpus/s34/text/" + index.split('.')[0] + ".txt"       # modify
    with open(name, 'w') as fw:
        fw.write(sentence)
```
